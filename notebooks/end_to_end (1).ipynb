```python
#Handling missing value
df["usia"] = df["usia"].fillna(df["usia"].median())
df["status_pekerjaan"] = df["status_pekerjaan"].fillna(df["status_pekerjaan"].mode()[0])
```

```python
df.isnull().sum()
```

```python
#Handling inconsistent Data

df["id_pelanggan"] = df["id_pelanggan"].str.upper()
df["id_pelanggan"] = df["id_pelanggan"].str.strip()

df["status_pekerjaan"] = df["status_pekerjaan"].str.lower()
df["status_pekerjaan"] = df["status_pekerjaan"].str.strip()

df["tipe_produk"] = df["tipe_produk"].str.lower()
df["tipe_produk"] = df["tipe_produk"].str.strip()

df["tujuan_pinjaman"] = df["tujuan_pinjaman"].str.lower()
df["tujuan_pinjaman"] = df["tujuan_pinjaman"].str.strip()
```

```python
#Cek Outlier
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns


for col in numerical_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(df[col])
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
cols = [
    "status_pinjaman",
    "gagal_bayar_tercatat",
    "tunggakan_2thn_terakhir",
    "catatan_negatif"
]

for col in cols:
    print(col)
    print(df[col].unique())
    print()
```

```python
# Handling outlier

numerical_cols = numerical_cols.drop([
    "status_pinjaman",
    "gagal_bayar_tercatat",
    "tunggakan_2thn_terakhir",
    "catatan_negatif"
], errors ="ignore")
for col in numerical_cols:
    df[col] = np.log1p(df[col])
    lower = df[col].quantile(0.01)
    upper = df[col].quantile(0.99)

    df[col] = df[col].clip(lower, upper)
```

```python
#Cek Outlier

for col in numerical_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(df[col])
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Feature Engineering

#Mengambil Hanya yang berkorelasi tinggi dengan target
fitur_pilihan = ["rasio_pinjaman_terhadap_pendapatan", "gagal_bayar_tercatat", "tunggakan_2thn_terakhir", "skor_kredit", "usia", "lama_riwayat_kredit_tahun"]

X = df[fitur_pilihan]
y = df["status_pinjaman"]

print(X.head())
print(y.head())


```

```python
# split DATA

from sklearn.model_selection import (
    train_test_split,
    GridSearchCV
)
X = df[fitur_pilihan]
y = df["status_pinjaman"]


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,  stratify=y, random_state=42)
```

```python
y.value_counts()
```

```python
#Preprocessing pipeline
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
preprocessing = Pipeline([('imputer', SimpleImputer(strategy='median'))])


```

```python
#Full Pipeline

pipeline = Pipeline([
    ('preprocessing', preprocessing),
    ('model', RandomForestClassifier(class_weight='balanced', random_state=42))
])
```

```python
#Train Model

pipeline.fit(X_train, y_train)
```

```python
#Prediksi

y_pred = pipeline.predict(X_test)
```

```python
#evaluasi

accuracy_score = pipeline.score(X_test, y_test)
print(f"Akurasi: {accuracy_score}")
```

```python
from sklearn.metrics import classification_report

# classification report
print(classification_report(y_test, y_pred))
```

```python
# cek apakah model overfitting / underfitting

train_score = pipeline.score(X_train, y_train)
test_score = pipeline.score(X_test, y_test)

print("Train Score :", train_score)
print("Test Score  :", test_score)
```

```python
#Hyperparameter Tuning

param_grid = {
    'model__n_estimators': [100, 200],
    'model__max_depth': [5, 10, None],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 5]
}



grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=3,
    scoring='f1',
    n_jobs=-1,
    verbose=2
)
```

```python
grid_search.fit(X_train, y_train)
```

```python
#Hasil setelah GridSearch

from sklearn.metrics import f1_score

best_model = grid_search.best_estimator_

print("Train:", f1_score(y_train, best_model.predict(X_train)))
print("Test :", f1_score(y_test, best_model.predict(X_test)))
```

```python

#Fungsi evaluasi percobaan
import optuna
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import Pipeline

def objective(trial):

    params = {
        "n_estimators": trial.suggest_int("n_estimators", 100, 500),
        "max_depth": trial.suggest_int("max_depth", 3, 30),
        "min_samples_split": trial.suggest_int("min_samples_split", 2, 20),
        "min_samples_leaf": trial.suggest_int("min_samples_leaf", 1, 10),
    }

    model = RandomForestClassifier(
        **params,
        random_state=42,
        n_jobs=-1
    )

    pipeline = Pipeline([
        ('model', model)
    ])

    score = cross_val_score(
        pipeline,
        X_train,
        y_train,
        cv=3,
        scoring='f1'
    ).mean()

    return score
```

```python
#menjalankan optuna
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=30)
```

```python
#Hasil optuna
print(study.best_value)
print(study.best_params)
```

```python
best_model = RandomForestClassifier(
    **study.best_params,
    random_state=42,
    n_jobs=-1
)

pipeline = Pipeline([
    ('preprocessing', preprocessing),
    ('model', best_model)
])

pipeline.fit(X_train, y_train)
```

```python
#Hasil Akhir

from sklearn.metrics import f1_score

y_train_pred = pipeline.predict(X_train)
train_score = f1_score(y_train, y_train_pred)
y_test_pred = pipeline.predict(X_test)
test_score = f1_score(y_test, y_test_pred)

print("Train F1:", train_score)
print("Test F1:", test_score)
```
